### Persebaran Data Numerik

```python
def spread_plot(data):
    name = data.name.upper()

    fig1, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={"height_ratios": (.25, .75)}, figsize=(12, 8))

    fig1.suptitle("Persebaran data {}".format(name), fontsize=16, fontweight='bold')

    sns.boxplot(x=data, ax=ax1, showmeans=True, color='blue', orient='h')
    ax1.set(xlabel='')

    sns.histplot(data, ax=ax2, kde=True, color="purple")
```

```python
numerical_cols = df.select_dtypes(include=np.number).columns

for col in numerical_cols:
    spread_plot(df[col])
```

Informasi dari plot persebaran data:
1. `Carat` miring ke kanan (right-skewed) serta memiliki banyak outliers di kanan
2. `Depth` terdistribusi normal, namun ada beberapa outliers di kanan dan kiri
3. `Table` cukup terdistribusi normal, namun ada banyak outliers di kanan dan kiri
4. `Price` sangat miring ke kanan (right-skewed), hanyak outliers di kanan, mayoritas data memiliki `Price` < 1500
5. `X` agak miring ke kanan, ada banyak outlier di kanan
6. `Y` terdistribusi normal, ada beberapa outliers di ujung kanan
7. `Z` terdistribusi normal, ada beberapa outliers di ujung kanan

### Analisis Data Kategorikal

```python
def percentage_plot(plot, feature):
    total = len(feature)

    for p in plot.patches:
        perc = '{:.1f}%'.format(100 * p.get_height()/total)

        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        plot.annotate(perc, (x, y), size=12)
```

```python
categorical_cols = df.select_dtypes(include=np.object).columns.to_list()

fig, ax2 = plt.subplots(1, 3, figsize=(20, 8))

for i in range(len(categorical_cols)):
    plt.figure(figsize=(12, 8))
    sns.countplot(x=categorical_cols[i], data=df, ax=ax2[i]).set(title=categorical_cols[i].upper())
    sns.despine(top=True, right=True, left=True, bottom=True)
    percentage_plot(ax2[i], df[categorical_cols[i]])
```

Informasi dari bar chart:
1. Sebagian besar berlian memiliki `Cut` ideal (40%)
2. 12.6% berlian memiliki `Color` terbaik (D), 5.2% memiliki `Color` paling buruk (J). Sebagian besar memiliki `Color` rata-rata (G)
3. Sebagian besar berlian memiliki `Clarity` di bawah rata-rata. Hanya ada 3.3% dengan `Clarity` terbaik dan 1.4% dengan `Clarity` terburuk

### Analisis Bivariate

### Perbandingan antar fitur

```python
def comparison_plot(data, target):
    plt.figure(figsize=(12, 8))

    num_cols = data.select_dtypes(include=np.number).columns.to_list()
    
    for i in range(len(num_cols)):
        plt.subplot(4, 2, i+1)
        sns.boxplot(x=data[target], y=data[num_cols[i]], data=data, orient='v').set(title=target.upper() + ' vs ' + num_cols[i].upper())
        sns.despine(top=True, right=True, left=True)
        plt.tight_layout()
```

#### Cut vs semua fitur


```python
comparison_plot(df, 'cut')
```

Informasi:
1. Fair `Cut` memiliki `Carat` tertinggi
2. Fair `Cut` memiliki `Depth` tertinggi
3. Ideal `Cut` memiliki `Table` terendah
4. Premium `Cut` memiliki `Price` tertinggi
5. Nilai `X` dari Fair dan Premium `Cut` adalah yang tertinggi
6. Nilai `Y` & `Z` dan `Cut` hampir tidak memiliki korelasi 

#### Color vs semua fitur

```python
comparison_plot(df, 'color')
```

Informasi:
1. Semakin baik `Color`, semakin rendah `Carat`
2. `Color` tidak ada memiliki korelasi dengan `Depth` dan `Table`
3. Semakin mahal berlian, semakin rendah kualitas `Color`
4. Nilai `X` meningkat seiring dengan peningkatan kualitas `Color`

#### Heatmap

Kita akan melihat hubungan antar fitur untuk menentukan fitur-fitur yang berpengaruh signifikan terhadap variabel target

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='Blues')
plt.show()
```

Informasi dari heatmap:
1. `Carat` memiliki hubungan kuat dengan `Price`
2. Semakin besar ukuran berlian, semakin tinggi `Price`-nya
3. `Table` dan `Price` hampir tidak relevan
4. `Depth` dan `Price` hampir tidak memiliki korelasi
